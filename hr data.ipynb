#### dropping the non requre values that are 'Gender','Business','Dependancies','Calls','Type','Billing','Rating','Base_pay','Unit_Price','low','Unit_Sales',  'Total_Sales','openingbalance','closingbalance','Volume','Bonus'])because their correlation with salary is more tha 80 percent
                      

```python
# Dropping the non required variables, more than 80 percent
dff = df.drop(columns=['Gender','Business','Dependancies','Calls','Type','Billing',
                       'Rating','Base_pay','Unit_Price','low','Unit_Sales',
                       'Total_Sales','openingbalance','closingbalance','Volume','Bonus'])
dff.head()
```

### label Encodinng ,converting categorical data into numerical data 

```python
#label encoding to convert cateograical data to into numerical data
dff['Education']=dff['Education'].map({'High School or less': 0, 'Intermediate': 1,'Graduation': 2, 'PG': 3})
#assigning high school =0,Intermediate =1 and gradaution=2
                                    
```

```python
dff.head() #head of the dataset after label encoding
```

## Handling Outliers

```python
Q1=dff.quantile(0.25).round(3)#taking lower quantile into Q2
Q3=dff.quantile(0.75).round(3)#taking upper quantile into Q3
IQR=Q3-Q1 #calculating inter quqnatile range
print(IQR) #printing quqantile range
```

```python
dfout=dff[~((dff<(Q1-1.5*IQR))|(dff>(Q3+1.5*IQR))).any(axis=1)] #removing outliers 
dfout.shape
```
